### Clean Data

```python
# drop columns
cols = ['WeightPounds','Fat','LogId']
df_weight.drop(columns=cols,inplace=True)
```

```python
# BMI = Weight (kg) / Height^2 (m2) 
# Height = sqrt(Weight/BMI)

df_weight['Height'] =  np.sqrt(df_weight['WeightKg']/df_weight['BMI']) * 100
df_weight
```

```python
# create id_dict

all_list = []

all_list.append(list(df_activity['Id'].unique()))
all_list.append(list(df_weight['Id'].unique()))
all_list.append(list(df_hr_s['Id'].unique()))
all_list.append(list(df_m['Id'].unique()))
all_list.append(list(df_h['Id'].unique()))

id_set = set().union(*all_list)
old_id_list = list(id_set)
old_id_list.sort()

new_id_list = range(1,len(old_id_list)+1,1)
new_id_list = list(new_id_list)
#print(new_id_list)

id_dict = dict((key, value) for (key, value) in zip(old_id_list, new_id_list))
```

```python
# reset id using id_dict

df_activity["Id"] = df_activity["Id"].map(id_dict)
df_weight["Id"] = df_weight["Id"].map(id_dict)
df_hr_s["Id"] = df_hr_s["Id"].map(id_dict)
df_m["Id"] = df_m["Id"].map(id_dict)
df_h["Id"] = df_h["Id"].map(id_dict)
```

### Edit Time

```python
# convert time columns to timedate

def to_datetime(x):
    return dt.datetime.strptime(x, '%m/%d/%Y')

def to_datetime_2(x):
    return dt.datetime.strptime(x, '%m/%d/%Y %I:%M:%S %p')
    
df_activity['timestamp'] = df_activity['timestamp'].apply(to_datetime)

df_weight['timestamp'] = df_weight['timestamp'].apply(to_datetime_2)
df_hr_s['timestamp'] = df_hr_s['timestamp'].apply(to_datetime_2)
df_m['timestamp'] = df_m['timestamp'].apply(to_datetime_2)
df_h['timestamp'] = df_h['timestamp'].apply(to_datetime_2)

df_activity['timestamp'] = df_activity['timestamp'].astype('datetime64[ns]') 
```

```python
def split_datetime(df): 

    df['year'] = df['timestamp'].dt.year
    df['month'] = df['timestamp'].dt.month
    df['day'] = df['timestamp'].dt.day
    
    df['dayweek'] = df['timestamp'].dt.day_name()

    return df

def split_datetime_2(df): 

    df['year'] = df['timestamp'].dt.year
    df['month'] = df['timestamp'].dt.month
    df['day'] = df['timestamp'].dt.day
    
    df['hour'] = df['timestamp'].dt.hour
    df['min'] = df['timestamp'].dt.minute
    df['sec'] = df['timestamp'].dt.second
    
    df['dayweek'] = df['timestamp'].dt.day_name()

    return df
```

```python
df_activity = split_datetime(df_activity)

df_weight = split_datetime_2(df_weight)
df_hr_s = split_datetime_2(df_hr_s)
df_m = split_datetime_2(df_m)
df_h = split_datetime_2(df_h)
```
