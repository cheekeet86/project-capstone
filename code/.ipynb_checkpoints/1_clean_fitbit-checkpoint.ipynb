### Merge Data

```python
# merge df_xxx_m -> df_m

cols = ['Id','ActivityMinute']

df_m = df_calories_m.merge(df_intense_m, on=cols)
df_m = df_m.merge(df_mets_m, on=cols)
df_m = df_m.merge(df_steps_m, on=cols)
```

```python
# merge df_xxx_h -> df_h

cols = ['Id','ActivityHour']

df_h = df_calories_h.merge(df_intense_h, on=cols)
df_h = df_h.merge(df_steps_h, on=cols)
```

### Clean Data

```python
# drop columns
cols = ['WeightPounds','Fat','LogId']
df_weight.drop(columns=cols,inplace=True)
```

```python
# BMI = Weight (kg) / Height^2 (m2) 
# Height = sqrt(Weight/BMI)

df_weight['Height'] =  np.sqrt(df_weight['WeightKg']/df_weight['BMI']) * 100
df_weight
```

### Edit Columns

```python
# use lowercase for column names

df_activity.columns = map(str.lower, df_activity.columns)
df_weight.columns = map(str.lower, df_weight.columns)
df_hr_s.columns = map(str.lower, df_hr_s.columns)
df_m.columns = map(str.lower, df_m.columns)
df_h.columns = map(str.lower, df_h.columns)
```

```python
# print column names

print(df_activity.columns)
print('')
print(df_weight.columns)
print('')
print(df_hr_s.columns)
print('')
print(df_m.columns)
print('')
print(df_h.columns)
```

```python
# rename columns

df_activity = df_activity.rename(columns={
                            'activitydate':'timestamp',
                            'totalsteps':'steps',
                            'totaldistance':'dist_total',
                            'trackerdistance':'dist_track', 
                            'loggedactivitiesdistance':'dist_log',   
                            'veryactivedistance':'dist_act_high',   
                            'moderatelyactivedistance':'dist_act_med',   
                            'lightactivedistance':'dist_act_low',
                            'sedentaryactivedistance':'dist_act_zero',
                            'veryactiveminutes':'mins_act_high',
                            'fairlyactiveminutes':'mins_act_med',
                            'lightlyactiveminutes':'mins_act_low',
                            'sedentaryminutes':'mins_act_zero',
                            })

df_weight = df_weight.rename(columns={
                            'date':'timestamp',
                            'weightkg':'weight',
                            'ismanualreport':'manual'                           
                            })

df_hr_s = df_hr_s.rename(columns={
                            'time':'timestamp',
                            'value':'hr',              
                            })

df_m = df_m.rename(columns={'activityminute':'timestamp'})

df_h = df_h.rename(columns={
                            'activityhour':'timestamp',
                            'totalintensity':'intensity_total',
                            'averageintensity':'intensity_avg',
                            'steptotal':'steps' 
                            })
```

```python
# print column names

print(df_activity.columns)
print('')
print(df_weight.columns)
print('')
print(df_hr_s.columns)
print('')
print(df_m.columns)
print('')
print(df_h.columns)
```
